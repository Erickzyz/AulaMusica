```python
# Inicializando a lista (tabela) de músicas
lista_musica = []

# --- PARTE 1: USANDO WHILE PARA PERGUNTAR ---
# Queremos 3 músicas, então vamos usar um contador
contador = 0

while contador < 3:
    nome = input(f"Digite o nome da música {contador + 1}: ")
    lista_musica.append(nome) # Adiciona a música na lista
    contador = contador + 1   # Aumenta o contador para não rodar infinitamente

print("-" * 30) # Apenas uma linha para separar no console

# --- PARTE 2: USANDO FOR PARA MOSTRAR ---
# Agora vamos percorrer a lista e mostrar cada item conforme o formato pedido
indice = 1
for musica in lista_musica:
    print(f"a musica {indice} é {musica}")
    indice = indice + 1
```
